```python
import psycopg2
import config as creds
import pandas as pd
from pathlib import Path
```

```python
def connect():
    
    # Set up a connection to the postgres server.
    conn_string = "host="+ creds.PGHOST +" port="+ "5432" +" dbname="+ creds.PGDATABASE +" user=" + creds.PGUSER \
                  +" password="+ creds.PGPASSWORD
    
    conn = psycopg2.connect(conn_string)
    print("Connected!")

    # Create a cursor object
    cursor = conn.cursor()
    
    return conn, cursor
```

```python

# Connecting to DB
conn, cursor = connect()

# SQL command to create inventory table
create_table = """
    CREATE TABLE IF NOT EXISTS life_expect (
    country VARCHAR(100),
    year_ int,
    status VARCHAR(40),
    life_expectancy Real,
    adult_mortality Real,
    infant_death int,
    alcohol real,
    percentage_expenditure Real,
    hepatitis Real,
    measles int,
    bmi real,
    under_five_deaths int,
    polio Real,
    total_expenditure real,
    diphtheria Real,
    hiv_aids real,
    gdp real,
    population_ real,
    thinness_1to19 real,
    thinness_5to9 real,
    income_composition real,
    schooling real,
    PRIMARY KEY (country, year_));
    """

# Execute SQL Command and commit to DB
cursor.execute(create_table)
conn.commit()

```
