# Population Data
### Life_Expectancy_Data had dirty values for population

```python
# Load the the dataset into the notebook
file_path = "PreClean/world_population.csv"
```

```python
# create the dataframe without the numbered rows by using index_col=0
df_population = pd.read_csv(file_path)
df_population.head(20)
```

```python
#drop excess columns and convert index
df_population_wide = df_population.drop(df_population.columns[[1,2,3]], axis=1).set_index(["Country Name"])
df_population_wide.head()
```

```python
#convert df from wide format to long format
year_list = list(df_population_wide.columns)
df_population_long = pd.melt(df_population_wide, value_vars = year_list, 
                        value_name ='Population', var_name = 'Year', ignore_index = False )
df_population_long.head(20)
```

```python
df_population_long.dtypes
```

```python
#Convert Year to INT
df_population_long["Year"] = pd.to_numeric(df_population_long["Year"])
```

```python
df_population_long.dtypes
```

```python
# Drop all rows outside of 2000-2015
df_population_2000_2015 = df_population_long.loc[
    (df_population_long['Year']>2004) & (df_population_long['Year']<2016)].sort_values(by = ['Country Name','Year'])
df_population_2000_2015.head(20)
```

```python
#Count Countries and NANs
print(df_population_2000_2015['Population'].isna().sum())

len(df_population_2000_2015.index.unique())
```

```python
#Find which Country has missing values
df_population_2000_2015[df_population_2000_2015['Population'].isna()]
```

```python
#Drop 'Not Classified Country Name' & Fill Eritrea 2011 Population for missing years
df_population_cleaned = df_population_2000_2015.loc[df_population_2000_2015.index != 'Not classified']
len(df_population_cleaned.index.unique())
eritrea_pop = df_population_cleaned['Population'].loc[(df_population_cleaned['Year'] == 2011) & 
                                                      (df_population_cleaned.index == 'Eritrea')]
df_population_cleaned['Population'] = df_population_cleaned['Population'].fillna(eritrea_pop)
```

```python
#Count Countries and NANs
print(df_population_cleaned['Population'].isna().sum())

len(df_population_cleaned.index.unique())
```
